```python
#Defining deflection due to udl
s=lambda q,L,E,I:((q*L*L*L*L)/(8*E*I)) #define lambda function
B=s(2.5,250,20,0.8196)# input values
print("the deflectioon due to udl:",B)
```

```python
##Defining deflection due to pointload A

D=lambda F,a,E,I,b:((F*a*a*a)/(8*E*I))*(1+(3*a/2*b))#define lambda function
C=D(100,250,20,0.8196,150)# input values
print("the deflectioon due to udl:",C)
```

```python
##Defining deflection due to pointload B

D1=lambda F,a,E,I,b:((F*a*a*a)/(8*E*I))*(1+(3*a/2*b))#define lambda function
E=D1(200,250,20,0.8196,150)# input values
print("the deflectioon due to udl:",E)
```
